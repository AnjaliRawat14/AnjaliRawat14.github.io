```python
#INSTALL NECESSARY LIBRARIES
import numpy as np
import matplotlib as plt
import seaborn as sns
import pandas as pd
```

```python
#IMPORT DATASET 

data = pd.read_csv(r"C:\Users\anjal\Downloads\SampleSuperstore.csv\SampleSuperstore.csv")

#r used for while importing csv just to adjust ["/" with "\"]

print(data)
```

```python
#HEAD AND TAIL 

#TOP 5 ROWS FROM THE DATASET
top = data.head()
print(top)

#BOTTOM 5 ROWS FROM THE DATASET
bottom = data.tail()
print(bottom)
```

```python
#Display basic information about the dataset like datatype and null values
detail = data.info()

#Display statistical summary of the dataset like count , mean,std ,min,max etc.
describe = data.describe()
print(describe)
```

```python
# Check for missing values
null = data.isnull().sum()
print(null)

# REMOVE MISSING VALUES
data.dropna(inplace= True)     #inplace used to add the modification permanently
print(data)

#Alternatively, fill missing values with a specific value
#1. backfill and frontfill
bfill = data.bfill(inplace=True)
print(bfill)
ffill = data.ffill()
print(ffill)

#IN OTHER SAY HANDLE THE MISSING VALUE IS TO FILL IT BY THE MEAN AND MODE OF THE COLUMN IF COLUMN HAVE CATEGORYICALL DATA THAN USE MODE ELSE USE MEAN/AVG
mean = data['Profit'].mean()
print(mean)

#the mean of profit column is 28.6 which is aaverage of whole column now if you add this to all the missing row the sum of whole column is same 
#before handling missing value and after handling missing value.
fillna = data['Profit'].fillna(28.6,inplace = True)
print(fillna)
```

```python
#check the duplicate rows
dup = data.duplicated().sum()
print(dup)

#drop duplicate rows
dropna = data.drop_duplicates(inplace = True)
print(dropna)
```
